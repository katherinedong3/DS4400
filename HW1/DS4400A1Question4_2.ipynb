```python
import numpy as np

rng = np.random.default_rng()

matrices = []
inverses = []

while len(matrices) < 3:
    M = rng.integers(-10, 11, size=(3, 3))
    if abs(np.linalg.det(M)) < 1e-9:
        continue
    matrices.append(M)
    inverses.append(np.linalg.inv(M))

for i, (M, Minv) in enumerate(zip(matrices, inverses), start = 1):
    print(f"\nMatrix",i,"\n", M)
    print(f"\nInverse of matrix #{i} (numpy):\n{Minv}")
    print(f"\nCheck M * (M inverse):\n{M @ Minv}")
```
